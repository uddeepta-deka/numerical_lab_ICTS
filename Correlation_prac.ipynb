```python
import numpy as np
import matplotlib.pyplot as plt
import math

alpha = 1.0
beta  = 0.05
#generating 1000 random integers between 0 and 50
x = np.random.randint(0,50,1000)
a = []
for item in x:
    a.append(math.sin(alpha*item + beta*item*item))

#positive correlation with some noise
y = a + np.random.normal (0,10,1000)

z = np.random.randint(0,50,1000)

plt.figure(figsize = (14,7))
plt.scatter(a,x)
plt.grid()

print(np.corrcoef(a,y))
plt.figure(figsize = (14,7))
plt.scatter(a,y)
plt.grid()

print(np.corrcoef(y,z))
plt.figure(figsize = (14,7))
plt.scatter(y,z)
plt.grid()

plt.figure(figsize = (14,7))
plt.xcorr(a,z,normed = True)
plt.show()
```
