# Stochastic ODE: Langevin equation

```python
import numpy as np                  #Using odeint to simulate the brownian motion
import scipy as sc
import matplotlib.pyplot as plt
import matplotlib
```

Brownian motion is described by the Langevin equation: $$m\frac{d^2\vec{x}}{dt^2}=-\lambda \frac{d\vec{x}}{dt} + \vec{\eta}(t)$$ where $m$ is the mass of the particle, $\vec{x}$ its position vector, $\lambda$ a damping constant and $\eta (t)$ (called the 'noise term') describes the stochastic process affecting the system.

```python
#verifying stochasticity by solving the equations for trajectories exactly
#defining the gaussian noise and returning xth element of the gaussuan noise
def noise(x):
    gaussian=np.random.normal(0,1,100) #Gaussian with 0 mean and unit variance
    return gaussian[x]


plt.figure(figsize = (16,8))     #we can still simulate the brownian by simple algorithm 
time   = np.arange(0,100,1) 
xn1    = np.zeros(100)
xn2    = np.zeros(100)

for i in range(len(time) - 1):
    xn1[i+1] = xn1[i] + noise(i)
    xn2[i+1] = xn2[i] + noise(i)     #computing the next point displacement of each particle

plt.plot(time,xn1,label='Particle 1')
plt.plot(time,xn2,label='Particle 2')
plt.xlabel('time',fontsize=15)
plt.ylabel('Displacement',fontsize=15)
plt.suptitle('Brownian Motion of two particles',fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

Brownian motion simulation for 1000 particles

```python
plt.figure(figsize = (16,8))
avg   = np.zeros(100)
xn    = np.zeros(100)

for j in range(1000):                 #particle index j
    noise2 = np.random.normal(0,1,100)
    for i in range(len(time)-1):      # i is time index
        xn[i+1] = xn[i] + noise2[i]
        avg[i+1] =+ xn[i+1]           # avg keeps total sum of ith displacement of 1000 particles
    plt.plot(time, xn, linewidth = 0.5)
    xn = np.zeros(100)   
avg = avg/1000
plt.xlabel('time',fontsize = 15)
plt.ylabel('Displacement',fontsize = 15)
plt.title("Brownian motion of Thousand Particles",fontsize=20)
plt.grid()

plt.figure(figsize = (16,8))
plt.plot(time, avg)
plt.xlabel('time',fontsize = 15)
plt.ylabel('Displacement',fontsize = 15)
plt.title("Average motion of Thousand Particles",fontsize=20)
plt.grid()

plt.show()
```
