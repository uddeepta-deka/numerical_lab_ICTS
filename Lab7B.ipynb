# Power Spectrum estimation using FFT

LIGO data from 2005 contains 256 seconds of data sampled at a rate of 4096 Hz. We need to compute the PSD (Power Spectral Density) of the data using Welch's modified periodogram method

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt 

#reading the data
data = np.loadtxt('/home/uddeepta/Desktop/ICTS/project_GW/computation/H2-STRAIN_4096Hz-815045078-256.txt')

fs   = 4096  #Sampling Frequency
sig  = np.asarray(data) #storing data as array

#Computing the power density by welches method
#This function returns array of sample frequencies and PSD of the signal
f, Pxx = signal.welch(sig, fs, nperseg=8192)    

#plotting the PSD
plt.figure(figsize=(16,8))
plt.semilogy(f, Pxx)
plt.xlim([0,2200])
plt.xlabel('frequency [Hz]', fontsize = 15)
plt.ylabel('PSD [Watt/Hz]', fontsize = 15)
plt.suptitle("Spectral Power Density Using Welch's Modified periodogram", fontsize = 20)
plt.show()
```
